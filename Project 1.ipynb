```python
# Import library
import cv2
import matplotlib.pyplot as plt
```

```python
# Membaca gambar
gambar = cv2.imread('citraBagas.jpg') # Ganti 'gambar.jpg' dengan nama file gambar yang ingin dianalisa
```

```python
# Konversi gambar ke ruang warna BGR
gambar_BGR = cv2.cvtColor(gambar, cv2.COLOR_RGB2BGR)
gambar_RGB = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)
```

```python
# Memisahkan kanal warna
kanal_merah = gambar_RGB[:,:,0]
kanal_hijau = gambar_RGB[:,:,1]
kanal_biru = gambar_RGB[:,:,2]
```

```python
# Menampilkan gambar kanal warna dalam skala abu-abu
plt.figure(figsize=(15, 4))

# Gambar Asli
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(gambar_RGB, cv2.COLOR_RGB2BGR))
plt.title('Gambar Asli')
plt.axis('off')

# Gambar Kanal Merah
plt.subplot(1, 4, 2)
plt.imshow(kanal_merah, cmap='gray')
plt.title('Kanal Merah')
plt.axis('off')

# Gambar Kanal Hijau
plt.subplot(1, 4, 3)
plt.imshow(kanal_hijau, cmap='gray')
plt.title('Kanal Hijau')
plt.axis('off')

# Gambar Kanal Biru
plt.subplot(1, 4, 4)
plt.imshow(kanal_biru, cmap='gray')
plt.title('Kanal Biru')
plt.axis('off')

plt.tight_layout()
plt.savefig('gambar_kanal_warna_dan_asli.png')
plt.show()
```

```python
# Menghitung histogram untuk setiap kanal warna
hist_merah = cv2.calcHist([kanal_merah], [0], None, [256], [0,256])
hist_hijau = cv2.calcHist([kanal_hijau], [0], None, [256], [0,256])
hist_biru = cv2.calcHist([kanal_biru], [0], None, [256], [0,256])

# Menampilkan histogram
plt.plot(hist_merah, color='r')
plt.plot(hist_hijau, color='g')
plt.plot(hist_biru, color='b')
plt.title('Histogram Warna Merah, Hijau, dan Biru')
plt.xlabel('Nilai Intensitas')
plt.ylabel('Frekuensi')
plt.savefig('histogram_warna.png')
plt.show()
```
